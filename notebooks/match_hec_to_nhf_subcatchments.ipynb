```python
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
hms_subbasins_path = '/home/akagi/Documents/Data/txswift/hms/Subbasin.geojson'
nhf_path = '/home/akagi/Documents/Data/txswift/nhf_1.1.3.gpkg'
```

```python
hms_subbasins = gpd.read_file(hms_subbasins_path)
nhf_subbasins = gpd.read_file(nhf_path, layer='divides')
```

```python
nhf_centroids = nhf_subbasins.copy()
nhf_centroids.set_geometry(nhf_subbasins.geometry.centroid, inplace=True)
nhf_centroids = nhf_centroids.to_crs(hms_subbasins.crs)
```

```python
nhf_hms_join = nhf_centroids.sjoin(hms_subbasins)
```

```python
assert not nhf_hms_join['div_id'].duplicated().any()
```

```python
id_mapping = nhf_hms_join[['div_id', 'name']].rename(columns={'div_id' : 'nhf_id', 'name' : 'hms_id'})
id_mapping
```

```python
id_mapping.to_csv('../data/nhf_to_hms_subbasins.csv')
```
